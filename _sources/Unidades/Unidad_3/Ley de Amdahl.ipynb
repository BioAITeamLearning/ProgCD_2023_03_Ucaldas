# Ley de Amdahl

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=np.linspace(1,100,100)
F1=0.1
F2=0.3
F3=0.5
F4=0.9
```

```python
S1=N/(N*F1+1-F1)
S2=N/(N*F2+1-F2)
S3=N/(N*F3+1-F3)
S4=N/(N*F4+1-F4)
```

```python
plt.plot(N,S1,N,S2,N,S3,N,S4)
plt.title("Ley Amdlah")
plt.xlabel("#Procesadores")
plt.ylabel("Speed-Up o Aceleraci√≥n")
plt.grid(True)
```

```python
E1=1/(N*F1+1-F1)
E2=1/(N*F2+1-F2)
E3=1/(N*F3+1-F3)
E4=1/(N*F4+1-F4)
```

```python
plt.plot(N,E1,N,E2,N,E3,N,E4)
plt.title("Ley Amdlah")
plt.xlabel("#Procesadores")
plt.ylabel("Eficiencia")
plt.grid(True)
```

```python
F=np.linspace(0,1,100)
N1=12
N2=100
N3=1000
```

```python
S1=N1/(N1*F+1-F)
S2=N2/(N2*F+1-F)
S3=N3/(N3*F+1-F)
```

```python
plt.plot(F,S1,F,S2,F,S3)
plt.title("Ley Amdlah")
plt.xlabel("Fraccion no paralelizable - F")
plt.ylabel("Speed-Up o Aceleraci√≥n")
plt.grid(True)
```
